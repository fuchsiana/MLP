# MLP regressor demo

### Import Packages Etc

```python
from IPython.display import display, HTML, Image

from MLP import MultiLayerPerceptron

import numpy as np

import math
```

### Generate Test Data

Generating 50 input vectors each containing 4 random number components, with a corresponding output of the following formula: 

For inputs [x1 x2 x3 x4], output = sin(x1-x2+x3-x4) 

```python
Input = np.random.normal(loc=0.0, scale=0.3, size=(50, 4))
print(Input)
```

```python
Output = np.array([[]], float)
for row in Input:
    temp = np.sin(row[0]-row[1]+row[2]-row[3])
    Output = np.append(Output, temp)
print(Output)
```

```python
Output = Output.reshape(50,1)
print(Output.shape)
print(Output)
```

### Training at an 80:20 train/test split

```python
# split data into test/training sets
XTest = Input[40:,:]
XTrain = Input[:40,:]

YTest = Output[40:,:]
YTrain = Output[:40,:]
```

```python
mlp_3 = MultiLayerPerceptron(hidden_units = 16, epochs = 100000, weights_range = 'xavier', \
                         batch_size = 8, random_seed = 23, verbose = 0)
mlp_3.fit(XTrain, YTrain)
```

```python
print("Test set actual outputs: \n", YTest)
XTest_predict = mlp_3.predict(XTest)

print()
print("Test set predictions: \n", XTest_predict)
```

```python
mse, mae, ae = mlp_3.calculate_error(YTest, XTest_predict, XTest_predict.shape[0])
print("MSE on test is", mse)
print("MAE on test is", mae)
```
